Ok so lets fix those types. Convert them to ints. If the type conversion fails, we now know we have further problems.

````python
# Columns we want to convert to integers
int_columns = ["year", "rating_count", "review_count"]

# Handle missing values and convert to integers
for col in int_columns:
    try:
        # Fill missing values with a placeholder or drop rows with NaNs
        df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to numeric, setting invalid values as NaN
        df = df[df[col].notnull()]  # Drop rows where conversion failed
        df[col] = df[col].astype(int)  # Convert to integer
    except Exception as e:
        print(f"Failed to convert column {col} to int: {e}")

# Verify the changes
df.info()
````

Lets wrap the above code into a function which we will then use

````python
def get_author(url):
    """
    Extracts and returns the author's name from the given author URL.

    Parameters:
        url (str): The author URL.

    Returns:
        str: The extracted author's name, or None if the URL format is invalid.
    """
    try:
        # Split the URL by '/' and extract the last part (or specific part containing the name)
        parts = url.split('/')
        # Assume the author's name is the last non-empty part of the URL
        author_name = parts[-1] if parts[-1] else parts[-2]
        # Replace hyphens with spaces to format the name properly
        author_name = author_name.replace('-', ' ')
        return author_name
    except Exception as e:
        print(f"Error processing URL {url}: {e}")
        return None
````

````python
# Example usage
sample_url = "https://www.goodreads.com/author/show/12345-Jane-Doe"
print(get_author(sample_url))  # Output: "Jane Doe"
````

````python
#Apply the get_author function to the 'author_url' column using '.map'
#and add a new column 'author' to store the names
df['author'] = df.author_url.map(get_author)
df.author[0:5]
````

<div class="exercise"><b>Exercise</b></div>

Write a function that accepts a genre url and returns the genre name based on your experimentation above



````python
def split_and_join_genres(url):
    """
    Extracts and returns genre names from a genre_url string.

    Parameters:
        url (str): The genre_url string containing pipe-separated genre URLs.

    Returns:
        str: A cleaned, pipe-separated string of genre names.
    """
    try:
        # Split the URL string by '|' to separate individual genre URLs
        genres = url.strip().split('|')
        # Extract and clean each genre name
        genre_names = [e.split('/')[-1].replace('-', ' ') for e in genres]
        # Join the cleaned genre names back into a single string
        return "|".join(genre_names)
    except Exception as e:
        print(f"Error processing genre URL {url}: {e}")
        return None
````

Lets get the best-rated book(s) for every year in our dataframe.

````python
# Group the dataframe by 'year' and find the book(s) with the highest rating for each year
best_books = df.loc[df.groupby('year')['rating'].idxmax()]

# Display the best-rated books for each year
print("Best-rated books for each year:")
best_books

# Optionally, save this to a new CSV file
best_books.to_csv("data/best_books_by_year.csv", index=False, header=True)
````
